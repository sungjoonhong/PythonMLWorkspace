```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('LinearRegressionData.csv')
```

```python
dataset.head()
```

```python
# dataset.iloc[row, col]
X = dataset.iloc[:, :-1].values # 처음부터 마지막 컬럼 직전까지의 데이터 (독립변수 - 원인)
y = dataset.iloc[:, -1].values # 마지막 컬럼 데이터 (종속 변수 - 결과)
```

```python
X, y
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression() # 객제 생성
reg.fit(X, y) # 학습 (모델 생성)
```

```python
y_pred = reg.predict(X) # X에 대한 예측 값
y_pred
```

```python
plt.scatter(X, y, color='blue') # 산점도
plt.plot(X, y_pred, color='green') # 선 그래프
plt.title('Score by hours')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
print('9시간 공부했을 때 예상 점수 : ', reg.predict([[9]])) # 2차원 배열 형태로 들어가야 함 # [9], [8], [7]
```

```python
reg.coef_ # 기울기 (m)
```

```python
reg.intercept_ # y 절편 (b)
```

### 데이터 세트 분리

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('LinearRegressionData.csv')
dataset
```

```python
X = dataset.iloc[:,:1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) #train:test = 80:20 으로 분리
```

```python
X, len(X) #전체 데이터 X, 개수
```

```python
X_train, len(X_train) # 훈련 세트 X, 개수
```

```python
X_test, len(X_test) # 테스트 세트 X, 개수
```

```python
y, len(y) # 전체 데이터 y
```

```python
y_train, len(y_train) # 훈련 세트 y
```

```python
y_test, len(y_test) # 테스트 세트 y
```

### 분리된 데이터를 통한 모델링

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train) # 훈련 세트로 학습
```

### 데이터 시각화 (훈련 세트)

```python
plt.scatter(X_train, y_train, color='blue') # 산점도
plt.plot(X_train, reg.predict(X_train), color='green') # 선 그래프
plt.title('Score by hours (train data)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```
